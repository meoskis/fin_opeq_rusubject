```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
# координаты по оси икс
x_axis = np.array([0., 0.18078584, 9.32526599, 17.09628721, 4.69820241, 11.57529305, 11.31769349, 14.63378951])

# Координаты по оси игрек
y_axis = np.array([0.0, 7.03050245, 9.06193657, 0.1718145, 5.1383203, 0.11069032, 3.27703365, 5.36870287])

shipments = np.array([5,7,4,3,5,2,1,1])

village = ['Тетерье','Журавец','Корсунь','Березовка','Протасово','Трудки','Нижний Туровец','Вышний Туровец']

data = pd.DataFrame({'x_coordinates_km': x_axis,'y_coordinates_km': y_axis,'deliveries':shipments}, index=village)
print(data)
```

```python
vectors = data[['x_coordinates_km', 'y_coordinates_km']].values
print(vectors)
```

```python
evclid_dist = pd.DataFrame(np.linalg.norm(vectors[:, np.newaxis] - vectors, axis=2), index=village, columns=village)
evclid_dist
```

```python
evclid_dist = evclid_dist * 2
```

```python
shipping_in_week = (evclid_dist * 2) * data['deliveries']
shipping_in_week
```

```python
shipping_in_week.drop('Тетерье', axis=0, inplace=True)
shipping_in_week.drop('Тетерье', axis=1, inplace=True)

# Находим населенный пункт с наименьшей суммарной дистанцией до соседей
optimal_location = shipping_in_week.sum(axis=1).idxmin()

print(f"Оптимальный населенный пункт для склада БПЛА: {optimal_location}")
```

```python
avenues_df = pd.DataFrame([0, 153, 307, 524], index=['Park', 'Lexington', '3rd', '2nd']) # Координаты авеню
streets_df = pd.DataFrame([0, 81, 159, 240, 324], index=['76', '75', '74', '73', '72']) # Координаты улиц

# Адрес это пересечение авеню и улицы
address = ['Lexington', '74']
# Адреса перекрестков на которых стоят такси
taxies = [
    ['Park', '72'],
    ['2nd', '75'],
    ['3rd', '76']
]
```

```python
taxies_distances = []
start = (avenues_df.loc[address[0]].values[0], streets_df.loc[address[1]].values[0],)

for taxi in taxies:
    taxi = (avenues_df.loc[taxi[0]].values[0], streets_df.loc[taxi[1]].values[0])
    taxies_distances.append(distance.cityblock(start, taxi))

taxies_distances = np.array(taxies_distances)
print("Порядковый номер", taxies_distances.argmin())
print("Перекресток", taxies[taxies_distances.argmin()])
```
